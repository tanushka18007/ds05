```python
import pandas as pd

csv_path = r"C:\Users\91887\OneDrive\Documents\Hate_Crimes_20240707.csv"
data = pd.read_csv(csv_path)

data.head(10)
```

```python
missing_values = data.isnull().sum()
print(missing_values)
data.fillna(data.mode().iloc[0], inplace=True)
missing_values_after = data.isnull().sum()
print(missing_values_after)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(22, 14))
sns.histplot(data['offense'], bins=24, kde=True)
plt.title('Distribution of Accidents over Offense')
plt.xlabel('offense')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(22, 14))
sns.countplot(y='victim_gender', data=data)
plt.title('Accidents by Road Conditions')
plt.xlabel('Number of Accidents')
plt.ylabel('Road Conditions')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='weekday', data=data)
plt.title('Accidents on which day')
plt.xlabel('Number of Accidents')
plt.ylabel('weekday')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='location_type', data=data)
plt.title('Location of accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Location of accidents')
plt.show()
```
